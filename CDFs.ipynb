```python
import xarray as xr
import numpy as np
import scipy.stats as ss
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Define functions
def open_ds(bubble, wind, hres, vres):
    filepath = (f'/chinook2/stluthi/cm1_output/{bubble}_bubble/{wind}/{hres}dx_{vres}/cm1out_{hres}dx_{vres}.nc')
    return xr.open_dataset(filepath)
```

```python
# Get SB NW datasets
cm1_3km50dz = open_ds('single', 'no_wind', '3km', '50vlvls')
cm1_3km100dz = open_ds('single', 'no_wind', '3km', '100vlvls')
cm1_1km50dz = open_ds('single', 'no_wind', '1km', '50vlvls')
cm1_1km100dz = open_ds('single', 'no_wind', '1km', '100vlvls')
```
